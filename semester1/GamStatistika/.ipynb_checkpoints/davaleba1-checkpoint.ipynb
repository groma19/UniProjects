```python
def createIntervals(array, num):
    maximum = max(array)
    minimum = min(array)
    interval = int((maximum - minimum) / num)
    
    categories = []
    
    tempNum = minimum
    
    for i in range(num - 1):
        categories.append((tempNum, tempNum + interval))
        tempNum += interval
    
    categories.append((tempNum, maximum+1))
    
    return categories
```

```python
intervals = createIntervals(numOfWords, 5)
```

```python
def intervalFrequency(intervals, array):
    intervalsFrequency = {}

    for i in range(len(intervals)):
        key = f"{intervals[i][0]}-{intervals[i][1]}"
        intervalsFrequency[key] = 0;

        for num in array:
            if ((num >= intervals[i][0]) & (num < intervals[i][1])):
                intervalsFrequency[key] = intervalsFrequency[key] + 1 
    
    return intervalsFrequency
```

```python
def intervalRelativeFrequency(intervalsFrequency, arrayLength):
    intervalsRelativeFrequency = {}
    
    for key, value in intervalsFrequency.items():
        intervalsRelativeFrequency[key] = value / arrayLength
    
    return intervalsRelativeFrequency
```

```python
intervalsFrequency = intervalFrequency(intervals, numOfWords)
```

```python
intervalsFrequency # ინტერვალები რაოდენობის მიხედვით
```

```python
def plotBar(placeholder, values, keys, ylabel, xlabel, title, useRotation=False):
    plt.bar(placeholder, values)
    
    if useRotation:
        plt.xticks(placeholder, keys, rotation=90)
    else:
        plt.xticks(placeholder, keys)
        
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    plt.title(title)
    
    plt.show()
```

```python
plotBar(np.arange(5), list(intervalsFrequency.values()), list(intervalsFrequency.keys()),
        "სიხშირე", "რაოდენობა", "სიხშირეთა განაწილება")
```

```python
intervalsRelativeFrequency = intervalRelativeFrequency(intervalsFrequency, len(numOfWords))
intervalsRelativeFrequency # ინტერვალები ფარდობითი განაწილების მიხედვით
```

```python
plotBar(np.arange(5), list(intervalsRelativeFrequency.values()), list(intervalsRelativeFrequency.keys()),
        "ფარდობითი სიხშირე", "რაოდენობა", "ფარდობით სიხშირეთა განაწილება")
```

```python
revenues = [
    10.6, 12.5, 14.6, 7.9, 10.8, 10.6, 6.5, 11.9,
    10.8, 6.0, 6.0, 5.9, 16.2, 13.3, 10.1, 9.0,
    14.8, 10.7, 12.8, 10.0, 18.4, 8.7, 8.7, 12.0,
    10.8, 11.0, 10.1, 10.6, 10.7, 15.4, 7.5, 9.1
]
```

```python
def relFreq(array, start, interval, useDelta=False):
    arrLength = len(array)
    minimum = start
    maximum = max(array) + 1
    
    categories = []
    
    if useDelta:
        delta = interval
        for i in range(int((maximum - minimum)/interval - 1)):
            categories.append((minimum, minimum + delta))
            minimum += delta
    
        categories.append((minimum, maximum))
    else:
        delta = int((maximum - minimum) / interval)
        
        for i in range(interval - 1):
            categories.append((minimum, minimum + delta))
            minimum += delta

        categories.append((minimum, maximum))
    
    
    intervalsFrequency = {}

    for i in range(len(categories)):
        key = f"{categories[i][0]}-{categories[i][1]}"
        intervalsFrequency[key] = 0

        for num in array:
            if ((num >= categories[i][0]) & (num < categories[i][1])):
                intervalsFrequency[key] = intervalsFrequency[key] + 1 
    
    intervalsRelativeFrequency = {}
    
    for key, value in intervalsFrequency.items():
        intervalsRelativeFrequency[key] = value / arrLength
    
    return intervalsRelativeFrequency
```

```python
relativeFrequencies = relFreq(revenues, 5.55, 2, True)
relativeFrequencies
```

a)

```python
plotBar(np.arange(len(list(relativeFrequencies.values()))), 
        list(relativeFrequencies.values()), 
        list(relativeFrequencies.keys()),
        "ფარდობითი სიხშირე", "ინტერვალი", "სააქციო საზოგადოების აქციათა მფლობელების განაწილება პროცენტის ფლობის მიხედვით", True)
```

b)

```python
plt.plot(list(relativeFrequencies.keys()), list(relativeFrequencies.values()), ".-b")
plt.xticks(rotation=90)
plt.ylabel("% სიხშირე")
plt.xlabel("კატეგორიები მოგების მიხედვით")
plt.title("აქციის მფლობელთა განაწილება მოგების მიხედვით")
plt.show()
```

```python
hydroMegavats = [
    12600, 10080, 10060, 6480, 6400, 6096, 6000, 5328, 5225, 4500,
    4150, 4050, 3600, 3575, 3409, 3300, 3200, 2700, 2700, 2700,
    2680, 2650, 2637, 2610, 2560, 2400, 2400, 2304, 2300, 2300,
    2250, 2124, 2100, 2100, 2069, 2031, 2030, 2000, 2000, 2000, 
    1979, 1824, 1807, 1800, 1800, 1800, 1800, 1750, 4600, 1950, 
    1890, 2100, 2100, 2500, 2416, 2715, 2820
]
```

```python
def roundToTenth(num):
    outNum = num
    
    if str(outNum)[-1] == "5":
        outNum += 1
    
    outNum = outNum / 10
    outNum = round(outNum)
    
    return outNum
```

```python
roundedMegavats = list(map(roundToTenth, hydroMegavats))
```

```python
def makeCategories(arr):
    categories = {}
    
    for num in arr:
        strNum = str(num)
        if strNum[:-2] not in categories:
            categories[strNum[:-2]] = [strNum[-2:]]
        else:
            categories[strNum[:-2]].append(strNum[-2:])
    
    return categories

roundedMegavatsCategories = makeCategories(roundedMegavats)
```

```python
def makeLeafDiagram(dict):
    for key in dict.keys():
        outStr = " " + str(key)
        outStr = outStr[-2:] + " |"
        for num in dict[key]:
            outStr += f" {num}"
        print(outStr)
```

```python
makeLeafDiagram(roundedMegavatsCategories)
```

b)

```python
def makeDoubleCategories(array):
    categories = makeCategories(array)
    print(categories)
    categoriesDoubled = {}
    for key in categories.keys():
        categoriesDoubled[key] = {
            "<4": [], 
            ">4": []
        }
        for num in categories[key]:
            if int(num[0]) < 4:
                categoriesDoubled[key]["<4"].append(num)
            else:
                categoriesDoubled[key][">4"].append(num)
    
    return categoriesDoubled    
```

```python
doubledCategories = makeDoubleCategories(roundedMegavats)
```

```python
def doubleLeafDiagram(doubleDict):
    for key in doubleDict.keys():
        strKey = " " + key
        strKey = strKey[-2:] + " |"
        for num in doubleDict[key]["<4"]:
            strKey += f" {num}"
        print(strKey)
        strKey = "   |"
        for num in doubleDict[key][">4"]:
            strKey += f" {num}"
        print(strKey)   
```

```python
doubleLeafDiagram(doubledCategories)
```
