```python
years = [1994, 1995, 1996, 1997, 1998]
salesHome = [250, 300, 350, 390, 400]
salesAll = [300, 350, 500, 520, 600]
```

```python
xLocs = np.arange(len(years))
width = 0.3

fig, ax = plt.subplots()

barsHome = ax.bar(xLocs - width/2, salesHome, width, label="საშინაო ბაზარზე", color="#67b1c8")
barsAll  = ax.bar(xLocs + width/2, salesAll, width, label="სულ" )

for bar in barsHome:
    bar.set_hatch("/")

ax.set_title("წარმოებული და გაყიდული პროდუქციის მოცულობები")
ax.set_ylabel("პროდუქციის მოცულობა (ათასობით ლარებში)")
ax.set_xticks(xLocs)
ax.set_xticklabels(years)
ax.legend()

plt.show()
```

```python
numOfWords = [205, 377, 292, 300, 179, 240, 300, 190, 680, 250,
              280, 170, 211, 266, 303, 350, 375, 288, 360, 225]
```

a) უწყვეტი

b)

```python
def createIntervals(array, num):
    maximum = max(array)
    minimum = min(array)
    interval = int((maximum - minimum) / num)
    
    categories = []
    
    tempNum = minimum
    
    for i in range(num - 1):
        categories.append((tempNum, tempNum + interval))
        tempNum += interval
    
    categories.append((tempNum, maximum+1))
    
    return categories
```

```python
intervals = createIntervals(numOfWords, 5)
```

```python
intervalsFrequency = {}

for i in range(len(intervals)):
    key = f"{intervals[i][0]} - {intervals[i][1]}"
    intervalsFrequency[key] = 0;
    
    for num in numOfWords:
        if ((num >= intervals[i][0]) & (num < intervals[i][1])):
            intervalsFrequency[key] = intervalsFrequency[key] + 1 
```

```python
intervalsFrequency # ინტერვალები რაოდენობის მიხედვით
```

```python
arrangeCategories = np.arange(5)
plt.bar(arrangeCategories, list(intervalsFrequency.values()))
plt.xticks(arrangeCategories, list(intervalsFrequency.keys()))
plt.ylabel("რაოდენობა")
plt.title("სიხშირეთა განაწილება")
plt.show()
```

```python
intervalsRelativeFrequency = {}
observedNum = len(numOfWords)

for key, value in intervalsFrequency.items():
    intervalsRelativeFrequency[key] = value / observedNum

intervalsRelativeFrequency # ინტერვალები ფარდობითი განაწილების მიხედვით
```

```python
plt.bar(arrangeCategories, list(intervalsRelativeFrequency.values()))
plt.xticks(arrangeCategories, list(intervalsRelativeFrequency.keys()))
plt.ylabel("ფარდობითი რაოდენობა")
plt.title("ფარდობით სიხშირეთა განაწილება")
plt.show()
```
