```python
sales_data        = [200, 400, 800, 1500, 3000]  # გაყიდვების მაჩვენებლები
employees_by_rank = [1, 4, 5, 3, 1]              # თანამშრომელთა რაოდენობა რანგების მიხედვით
```

```python
def salary_by_rank(sales, employees):
    fund_coeff   = 0.05
    golden_coeff = 1.618
    
    salary_fund = sales * fund_coeff
    print("სახელფასო ფონდი (ათასი ლარი): ", int(salary_fund * 1000))
    
    cumulative_coeff = 0
    for i in range(len(employees_by_rank)):
        cumulative_coeff += employees_by_rank[i] * (golden_coeff**i)
    
    minimum_wage = salary_fund / cumulative_coeff  
    
    wage_by_ranks  = {}
    wage_by_groups = {}
    print("ხელფასები რანგის ზრდის მიხედვით")
    for i in range(len(employees_by_rank)):
        wage_by_ranks[i]  = minimum_wage * (golden_coeff**i)
        wage_by_groups[i] = minimum_wage * employees_by_rank[i] * (golden_coeff**i)
        print(f"რანგი {i + 1}: ", int(wage_by_ranks[i]*1000), f"--> ჯამური ხელფასი მოცემული რანგისთვის", int(wage_by_groups[i]*1000))
    
    print("-----------------------------")
    return wage_by_ranks, wage_by_groups
```

```python
for i in range(len(sales_data)):
    by_ranks, by_groups = salary_by_rank(sales_data[i], employees_by_rank)
```
